```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import AdaBoostClassifier
import talib
%matplotlib inline
pd.set_option('display.max_rows',10)
```

```python
contents = 'I:/data/Wind/Machine Learning/'
data_name = 'RU.csv'
df = contents+data_name
df = pd.read_csv(df)
df.fillna(0,inplace=True)
df.drop('ticker',axis=1,inplace=True)
df.set_index('TimeStamp',inplace=True)
print(df.shape)
columns_name = df.columns
columns_name1 = ['open', 'high', 'low', 'close', 'volume', 'amt', 'pct_chg']
columns_name2 =  ['PVC','SKW', 'KTS', 'SGH', 'SR', 'VOL', 'CSHSKW', 'CSHKTS']
def add_features(data,feature_list,n):
    
    for i in range(n):
        for j in feature_list:
            if j in data.columns:
                data['{}_{}'.format(j,i+1)]=data[j].shift(i+1)
            else:
                print('feature not in data')
                
def add_labels(data,threshold1=0,threshold2=0):
    if 'labels' in data.columns:
        data.drop('labels',axis=1,inplace=True)
    data['labels']=0
    data.loc[data['pct_chg_future']>threshold1,'labels']=1
    data.loc[data['pct_chg_future']<threshold2,'labels']=-1
    
def data_split(data,train_size,suffle=False):
    
    train_num = int(train_size * data.shape[0])
    if not suffle:
        
        train_data = data[:train_num]
        train_x = train_data.drop(['pct_chg_future','F_low','F_high','labels'],axis=1)
        train_y = train_data[['labels']]
        
        test_data = data[train_num:]
        test_x = test_data.drop(['pct_chg_future','F_low','F_high','labels'],axis=1)
        test_y = test_data[['labels']]
    else:
        X = data.drop(['pct_chg_future','F_low','F_high','labels'],axis=1)
        y = data[['labels']]
        train_x,test_x,train_y,test_y = train_test_split(X,y,test_size=1-train_size,random_state=42)
    return train_data,test_data,train_x,test_x,train_y,test_y
df['MA5']=df['pct_chg'].rolling(5).mean()
df['MA10']=df['pct_chg'].rolling(10).mean()
df['MA20']=df['pct_chg'].rolling(20).mean()
df['MA40']=df['pct_chg'].rolling(40).mean()
```

```python
yuzhi_list = np.arange(0.0002,0.004,0.00005) 
nums = len(yuzhi_list)
rtn=np.zeros(nums)
rtn_actual=np.zeros(nums)
std=np.zeros(nums)
SR=np.zeros(nums)
for i in range(nums): 
    data = df.copy()
    add_features(data,columns_name1,4)
    add_features(data,columns_name2,4)
    data['pct_chg_future']=data['pct_chg'].shift(-1)
    data['F_low']=data['low'].shift(-1)
    data['F_high']=data['high'].shift(-1)
    data.dropna(inplace=True)
    add_labels(data,yuzhi_list[i],-yuzhi_list[i])
    train_data,test_data,train_x,test_x,train_y,test_y = data_split(data,0.8,suffle=False)
    print('data size:',data.shape)
    print()
#     print(train_y[train_y.labels==0].shape[0]+test_y[test_y.labels==0].shape[0])
#     print(data[abs(data['pct_chg_future'])<=0.0004].shape[0])

    model = LogisticRegression()
    model.fit(train_x,train_y)
    y_pred = model.predict(test_x)
    print(model.score(train_x,train_y),model.score(test_x,test_y))
    test_y['y_pred'] = y_pred
    print(test_y['y_pred'].value_counts())
    test_data['y_pred'] = y_pred


    test_data['action']=test_data['y_pred']
    test_data.loc[(test_data['action']==1)&(test_data['F_low']>test_data['close']),['action']]=0
    test_data.loc[(test_data['action']==-1)&(test_data['F_high']<test_data['close']),['action']]=0
    test_data.action.value_counts()

    review = test_data[['pct_chg_future','action']]
    rtn[i]=(review['pct_chg_future']*review['action']).mean()*252*24
    
    buy_sell_fee = 0.0004
    review['fee']=abs(review['action']-review['action'].shift(1))*buy_sell_fee
    rtn_actual[i]=(review['pct_chg_future']*review['action']-review['fee']).mean()*252*24
    std[i]=(review['pct_chg_future']*review['action']).std()*((252*24)**0.5)
    SR[i]=(rtn_actual[i])/std[i]
#     print('Return with no transaction fee:',rtn)
#     print('Return after transaction fee:',rtn_actual)
#     print('Standard deviation:',std)
#     print('Sharp Ratio:',Sharp_Ratio)
#     print('N:',review.shape[0])
```

```python
plt.figure(figsize=(15,15))
plt.subplot(221)
plt.plot(yuzhi_list,rtn)
plt.xlabel('yuzhi_list')
plt.ylabel('Return without transaction fee')
plt.subplot(222)
plt.plot(yuzhi_list,rtn_actual)
plt.xlabel('yuzhi_list')
plt.ylabel('Return with transaction fee')
plt.subplot(223)
plt.plot(yuzhi_list,std)
plt.xlabel('yuzhi_list')
plt.ylabel('Standard Deviation')
plt.subplot(224)
plt.plot(yuzhi_list,SR)
plt.xlabel('yuzhi_list')
plt.ylabel('Sharp Ratio')
plt.show()
```
